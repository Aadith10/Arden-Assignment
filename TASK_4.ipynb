```python
import numpy as np
from scipy import stats
data = np.array([1050, 1070, 1065, 1080, 1045, 1075, 1060, 1070])
n = len(data)
mean = np.mean(data)
std_dev = np.std(data, ddof=1)
alpha = 0.05
df = n - 1
t_crit = stats.t.ppf(1 - alpha/2, df)
se = std_dev / np.sqrt(n)
margin = t_crit * se
lower = mean - margin
upper = mean + margin

print("Task 4 - 95% Confidence Interval for District Populations")
print("=" * 60)
print(f"Sample size (n):           {n}")
print(f"Sample mean:               {mean:.2f} residents")
print(f"Sample std dev (s):        {std_dev:.2f} residents")
print(f"Degrees of freedom:        {df}")
print(f"t-critical value (0.025):  {t_crit:.4f}")
print(f"Standard error (SE):       {se:.2f}")
print(f"Margin of error (ME):      {margin:.2f}")
print("=" * 60)
print(f"95% Confidence Interval:   ({lower:.2f}, {upper:.2f})")
print("=" * 60)
```
