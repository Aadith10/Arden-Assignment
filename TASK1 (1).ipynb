```python
import numpy as np

A = np.array([[1.2, 0.1],
              [0.05, 1.1]])

eigvals, eigvecs = np.linalg.eig(A)

idx = np.argsort(eigvals.real)[::-1]
eigvals = eigvals[idx]
eigvecs = eigvecs[:, idx]

v1 = eigvecs[:, 0].real
v1 = v1 / v1[1]
v2 = eigvecs[:, 1].real
v2 = v2 / v2[1]

l1 = round(eigvals[0].real, 3)
l2 = round(eigvals[1].real, 3)
v1_0 = round(v1[0], 3)
v2_0 = round(v2[0], 3)

print("\nEigenvalues")
print(f"Eigenvalue for λ1 ≈ {l1}")
print(f"Eigenvalue for λ2 ≈ {l2}")

print("\nEigenvectors")
print(f"Eigenvector for v1 ∝ ({v1_0}, 1)")
print(f"Eigenvector for v2 ∝ ({v2_0}, 1)")
```
